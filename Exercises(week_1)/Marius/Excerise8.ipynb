```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
import scipy
```

# Part 1

```python
# Parameters
rng = np.random.default_rng(30)
X = np.array([56, 101, 78, 67, 93, 87,64, 72, 80, 69])
a = -5
b = 5
n = len(X)


xbar = np.mean(X)
print("sample mean =", xbar)


B = 10_000
sum = 0
for bi in range(B):
    sample = rng.choice(X, size=len(X), replace=True)

    mean_star = np.mean(sample)
    obs = mean_star-xbar

    if ((a < obs) and (obs < b)):
        sum += 1

    

p_estimate = 1/B * sum
print(f"P_estimate: {p_estimate}")
```

# Part 2

```python
# Parameters
rng = np.random.default_rng(30)
X = np.array([5,4,9,6,21,17,11,20,7,10,21,15,13,16,8])
n = 15

xbar = np.mean(X)
print("sample mean =", xbar)


B = 10_000
VARS = []
for bi in range(B):
    sample = rng.choice(X, size=n, replace=True)

    s2_star = np.var(sample, ddof=1)
    VARS.append(s2_star)

    
print(f"VAR[s^2]: {np.var(VARS, ddof=1)}")
```

# Part 3

```python
# Parameters

def program(X, k=100):
    rng = np.random.default_rng(30)
    n = len(X)

    sample_mean = np.mean(X)
    sample_median = np.median(X)

    MEANS = np.empty(k)
    MEDIANS = np.empty(k)
    for i in range(k):
        sample = rng.choice(X, size=n, replace=True)
        
        MEANS[i] = np.mean(sample)
        MEDIANS[i] = np.median(sample)


    mean_variance_estimate = np.var(MEANS, ddof=1)
    median_variance_esimate = np.var(MEDIANS, ddof=1)

    return mean_variance_estimate, median_variance_esimate
```

```python
rng = np.random.default_rng(30)
n = 200
k = 1.05
beta = 1

true_mean = (k*beta)/(k-1)
true_median = beta*2**(1/k)
print(f"True mean: {true_mean}")
print(f"True median: {true_median}")
X = (rng.pareto(k, size=n) + 1) * beta
program(X)
```
