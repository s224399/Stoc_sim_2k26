```python
import numpy as np
import scipy.stats as ss
import scipy as sp
import matplotlib.pyplot as plt 
import time
```

```python
def sim_exp(lam,N):
    return -np.log(np.random.uniform(0,1,size=N))/lam


def sim_normal(N):
    n_pairs = (N + 1) // 2
    U1 = np.random.uniform(0, 1, size=n_pairs)
    U2 = np.random.uniform(0, 1, size=n_pairs)

    R = np.sqrt(-2 * np.log(U1))
    Z1 = R * np.cos(2 * np.pi * U2)
    Z2 = R * np.sin(2 * np.pi * U2)

    z = np.concatenate([Z1, Z2])
    return z


def sim_pareto(beta, k, N):
    U = np.random.uniform(0, 1, size=N)
    return beta * (U**(-1/k) - 1)


```
