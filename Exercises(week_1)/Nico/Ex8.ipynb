```python
import numpy as np
import matplotlib.pyplot as plt
```

### Part 1

```python
seed = 30
rng = np.random.default_rng(seed=seed)

X = np.array([56,101,78,67,93,87,64,72,80,69])
a = -5
b = 5
n = len(X)
mu = np.mean(X)
print(f'Estimated mean from original sample: {mu}')

k = 1_000
count = 0

for i in range(k):
    sample = rng.choice(a=X, size=n, replace=True)
    sample_mean = np.mean(sample)
    if(a < sample_mean-mu and sample_mean-mu < b):
        count += 1
P = count/k
print(f'Estimated P from {k} bootstraps: {P}')
```

### Part 2

```python
seed = 30
rng = np.random.default_rng(seed=seed)

X = np.array([5,4,9,6,21,17,11,20,7,10,21,15,13,16,8])
n = len(X)

k = 1_000
S2 = np.zeros(k)

for i in range(k):
    sample = rng.choice(a=X, size=n, replace=True)
    S2[i] = np.var(sample, ddof=1)

res = np.var(S2, ddof=1)
res
```

### Part 3

```python
def bootstrap(X, k, rng):
    n = len(X)
    mean = np.mean(X)
    median = np.median(X)

    means = np.zeros(k)
    medians = np.zeros(k)

    for i in range(k):
        sample = rng.choice(a=X, size=n, replace=True)
        means[i] = np.mean(sample)
        medians[i] = np.median(sample)

    mean_var = np.var(means, ddof=1)
    median_var = np.var(medians, ddof=1)

    print(f'Sample mean: {mean}')
    print(f'Estimated variance of sample mean: {mean_var}')
    print(f'Sample median: {median}')
    print(f'Estimated variance of sample median: {median_var}')
```

```python
seed = 30
rng = np.random.default_rng(seed=seed)
n = 200
beta = 1
X = (rng.pareto(a=1.05, size=n)+1)*beta

bootstrap(X=X, k=100, rng=rng)
```
