```python
np.random.seed(30)

a = 1.05
n = 200
beta= 1
real = (np.random.pareto(a,size = 200) +1)*beta


sample_mean = np.mean(real)
sample_median =np.median(real)

sample_mean,sample_median
```

```python
np.max(real)
```

```python
X= real

N = 100

mean_arr= np.zeros(N)
var_median= np.zeros(N)
for i in range(N):
    Xs = np.mean(np.random.choice(X, size=n, replace=True))
    mean_arr[i] = Xs
    Xs = np.median(np.random.choice(X, size=n, replace=True))
    var_median[i] = Xs


sample_mean_variance_bootstrap = np.var(mean_arr,ddof=1)
sample_median_variance_bootstrap = np.var(var_median,ddof=1)


bootstrap_mean = np.mean(mean_arr),
bootstrap_median = np.mean(var_median)

bootstrap_mean,bootstrap_median
```
