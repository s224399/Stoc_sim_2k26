```python
import math
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

rng = np.random.default_rng(30)
```

Part 1

```python
def bootstrap_p(X, r, a, b, rng):
    n = len(X)
    x_bar = np.mean(X)      # Observed sample mean (used as estimate of mu)
    means = np.zeros(r)

    for i in range(r):
        sample = rng.choice(X, size=n, replace=True)
        means[i] = np.mean(sample)

    diff = means - x_bar
    p = np.mean((diff > a) & (diff < b))

    return p
```

```python
X = [56, 101, 78, 67, 93, 87, 64, 72, 80, 69]
r = 10000       # Number of bootstrap replicates
a, b = -5, 5
rng = np.random.default_rng(30)
p = bootstrap_p(X, r, a, b, rng)
print("We estimate p =", p)
```

Part 2

```python
def bootstrap_var(X, r, rng):
    n = len(X)
    vars = np.zeros(r)

    for i in range(r):
        sample = rng.choice(X, size=n, replace=True)
        vars[i] = np.var(sample)

    var_boot = np.var(vars, ddof=1)

    return var_boot
```

```python
X = [5, 4, 9, 6, 21, 17, 11, 20, 7, 10, 21, 15, 13, 16, 8]
r = 10000       # Number of bootstrap replicates
rng = np.random.default_rng(30)
var_boot = bootstrap_var(X, r, rng)
print("Bootstrap estimate of Var(S^2):", var_boot)
```

Part 3

```python
def bootstrap_var2(X, r, mean_or_median, rng):
    n = len(X)
    vars = np.zeros(r)

    for i in range(r):
        sample = rng.choice(X, size=n, replace=True)
        vars[i] = mean_or_median(sample)

    mean_boot = np.mean(vars)
    var_boot = np.var(vars, ddof=1)

    return mean_boot, var_boot
```

```python
# Generate the Pareto sample

rng = np.random.default_rng(30)

n = 200
beta = 1
k = 1.05

X_pareto = (rng.pareto(k, n) + 1) * beta
```

```python
# Step 1: Compute the sample mean and sample median.

sample_mean = np.mean(X_pareto)
sample_median = np.median(X_pareto)
print("Sample mean:", sample_mean)
print("Sample median:", sample_median)
```

```python
# Step 2: Estimate the variance of the sample mean using the bootstrap method.

rng = np.random.default_rng(30)

k = 100
mean_boot_mean, var_mean_boot = bootstrap_var2(X_pareto, k, np.mean, rng)

print("Mean:", mean_boot_mean)
print("Variance (mean):", var_mean_boot)
```

```python
# Step 3: Estimate the variance of the sample median using the bootstrap method.

rng = np.random.default_rng(30)

k = 100
median_boot_mean, var_median_boot = bootstrap_var2(X_pareto, k, np.median, rng)

print("Median:", median_boot_mean)
print("Variance (median):", var_median_boot)
```

```python
# Step 4: Compare the precision of the two estimators.

print("\nComparison:")
print("Variance of mean   :", var_mean_boot)
print("Variance of median :", var_median_boot)
```
