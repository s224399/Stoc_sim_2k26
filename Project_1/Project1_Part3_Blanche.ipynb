```python
rng = np.random.default_rng(30)

Q = np.array([[-0.0085, 0.005, 0.0025, 0,     0.001],
              [0,      -0.014, 0.005,  0.004, 0.005],
              [0,       0,    -0.008,  0.003, 0.005],
              [0,       0,     0,     -0.009, 0.009],
              [0,       0,     0,      0,     0]])

n = 1000
death = 4
reappar = 2
lifetimes = np.zeros(n)
has_reappeared = np.zeros(n)
Y = []

for i in range(n):
    t = 0
    state = 0
    next_obs = 0 # Observation times (0, 48, 96, ...)
    X = []

    while state != death:
        t += rng.exponential(-1 / Q[state, state])

        while next_obs <= t:
            X.append(int(state))
            next_obs += 48

        proposals = [0,1,2,3,4]
        proposals.pop(state)

        probability = - (Q[state,proposals]) / (Q[state,state])
        state = rng.choice(a=proposals, p=probability)

        if (state == reappar and t > 30.5):
            has_reappeared[i] = 1
        
        if (state == death):
            lifetimes[i] = t
            X.append(death)
            break

    Y.append(X)

Y
```

```python
rng = np.random.default_rng(30)

Q = np.array([[-0.01, 0.0025, 0.0025, 0.0025, 0.0025],
               [ 0,   -0.0075, 0.0025, 0.0025, 0.0025],
               [ 0,    0,     -0.0050, 0.0025, 0.0025],
               [ 0,    0,      0,     -0.0025, 0.0025],
               [ 0,    0,      0,      0,      0     ]])

Q_prev = np.zeros(5)


while np.max(np.abs(Q - Q_prev)) >= 1e-3:

    for i in range(n):
        t = 0
        state = 0
        next_obs = 0 # Observation times (0, 48, 96, ...)
        X = []

        while state != death:
            t += rng.exponential(-1 / Q[state, state])

            while next_obs <= t:
                X.append(int(state))
                next_obs += 48

            proposals = [0,1,2,3,4]
            proposals.pop(state)

            probability = - (Q[state,proposals]) / (Q[state,state])
            state = rng.choice(a=proposals, p=probability)

            if (state == reappar and t > 30.5):
                has_reappeared[i] = 1
            
            if (state == death):
                lifetimes[i] = t
                X.append(death)
                break

    Y.append(X)

    N = np.zeros((5, 5))
    S = np.zeros(5)
    for x in Y:
        for t in range(len(x) - 1):
            i = x[t]
            j = x[t + 1]
            S[i] += 48
            if i != j:
                N[i, j] += 1

    Q_prev = Q.copy()

    for i in range(4):
        if S[i] > 1e-8:
            for j in range(4):
                if i != j:
                    Q[i][j] = N[i][j] / S[i]
            Q[i,i] = - np.sum(Q[i])
        else:
            Q[i, :] = 0
            Q[i, i] = 0

    print(Q)
    
    
```
